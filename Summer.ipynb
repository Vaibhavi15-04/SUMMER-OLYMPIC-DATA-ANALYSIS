**Summer Olympics Data Analysis Assignment**

```python
import pandas as pd
```

```python
df=pd.read_csv('summer.csv')
```

**1. In how many cities Summer Olympics is held so far??**


```python
len(df['City'].unique())
```

**2. Which sport is having most number of Gold Medals so far? (Top 5)**

```python
medal=df['Medal']=='Gold'
sport_gold=df[medal]
sport=sport_gold.groupby('Sport').count()['Medal'].sort_values(ascending = False)
sport.head().plot.bar(figsize = (5,5))
```

**3. Which sport is having most number of medals so far? (Top 5)**

```python
df.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (15,5))
```

**4. Which player has won most number of medals? (Top 5)**

```python
athlete=df.groupby('Athlete').count()['Medal'].sort_values(ascending = False)

athlete.head().plot.bar()
```

**5. Which player has won most number Gold Medals of medals? (Top 5)**

```python
gold=df['Medal']=='Gold'
athlete_gold=df[gold]
player=athlete_gold.groupby('Athlete').count()['Medal'].sort_values(ascending = False)
player.head().plot.bar(figsize = (5,5))
```

**6. In which year India won first Gold Medal in Summer Olympics?**

```python
country=df['Country']=='IND'
india=df[country]
gold=india['Medal'] == 'Gold'
gold_medal=india[gold]
year=gold_medal.head(1)
year['Year']
```

**7. Which event is most popular in terms on number of players? (Top 5)**

```python
df.groupby('Discipline').count()['Athlete'].sort_values(ascending = False).head().plot.bar(figsize = (10,5))
```

**8. Which sport is having most female Gold Medalists? (Top 5)**

```python
b=df['Gender']=='Women'
female=df[b]
gold=female['Medal'] == 'Gold'
gold_medal=female[gold]
gold_medal


gold_medal.groupby('Sport').count()['Medal'].sort_values(ascending = False).head().plot.bar(figsize = (15,5))
```
